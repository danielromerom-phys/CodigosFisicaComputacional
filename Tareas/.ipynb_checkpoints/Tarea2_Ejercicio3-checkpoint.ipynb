```python
# Importación de bibliotecas
import numpy as np
import math as mt
```

```python
# Variables Globales

a1 = 15.8
a2 = 18.3
a3 = 0.714
a4 = 23.2
```

```python
# Función para el inciso a) que toma de entrada valores de A y Z y retornna la energía de enlace B del átomo correspondiente.

def FSM_B_a(A, Z):
    
    if (A%2 != 0):
        a5 = 0
    elif (A%2 == 0 & Z%2 == 0):
        a5 = 12
    elif (A%2 == 0 & Z%2 != 0):
        a5 = -12

    B = (a1*A) - (a2*(A**(2/3))) - ((a3*(Z**2))/(A**(1/3))) - ((a4*((A-2*Z)**2))/A) + (a5/(A**(1/2)))

    print("La energía de enlace del átomo es de:", B, "MeV")
```

```python
FSM_B_a(58, 28)
```

```python
# Función para el inciso b) que toma de entrada valores de A y Z y retornna la energía de unión por nucleón, que es B/A.

def FSM_B_b(A, Z):
    
    if (A%2 != 0):
        a5 = 0
    elif (A%2 == 0 & Z%2 == 0):
        a5 = 12
    elif (A%2 == 0 & Z%2 != 0):
        a5 = -12

    B = (a1*A) - (a2*(A**(2/3))) - ((a3*(Z**2))/(A**(1/3))) - ((a4*((A-2*Z)**2))/A) + (a5/(A**(1/2)))
    
    print("La energía de unión por nucleón del átomo es de: ", B/A, "MeV por nucleón.")
```

```python
FSM_B_b(58, 28)
```

```python
# Función para el inciso c) que pasa por A=Z hasta A=3Z, compara el que tiene mayor energía de enlace por nucleón 
# y retorna el valor de A y la energía de enlace por nucleón B/A de dicho átomo con mayor energía de enlace por nucleón.

def FSM_B_c(Z):

    EUN = []
    A = [i for i in range(Z,(3*Z)+1,1)]

    for j in range(0, len(A)):
        if (A[j]%2 != 0):
            a5 = 0
        elif (A[j]%2 == Z%2 == 0):
            a5 = 12
        elif ((A[j]%2 == 0) & (Z%2 != 0)):
            a5 = -12
        
        B = (a1*A[j]) - (a2*(A[j]**(2/3))) - ((a3*(Z**2))/(A[j]**(1/3))) - ((a4*((A[j]-2*Z)**2))/A[j]) + (a5/(A[j]**(1/2)))

        EUN.append(B/A[j])

    for k in range(0, len(EUN)):
        if EUN[k] == np.max(EUN):
            A_max = A[k]
        else:
            continue

    print("El número de masa con el cual el átomo tiene mayor energía de enlace por nucleón es de:", A_max)
    print("La mayor energía de enlace por nucleón de éste átomo es de:", np.max(EUN), "MeV")
```

```python
FSM_B_c(28)
```

```python
# Función para el inciso d) que toma de entrada el valor Z, pasa por Z=1 hasta Z=100, asigna valores de A=Z hasta A=3Z, compara el que tiene
# mayor energía de enlace por nucleón y retorna el valor de A y la energía de enlace por nucleón B/A de dicho átomo con mayor energía de 
# enlace por nucleón.

def FSM_B_d():

    EUN_max = []

    for Z in range(1, 101, 1):
        EUN = []
        A = [i for i in range(Z,(3*Z)+1,1)]
    
        for j in range(0, len(A)):
            if (A[j]%2 != 0):
                a5 = 0
            elif (A[j]%2 == Z%2 == 0):
                a5 = 12
            elif ((A[j]%2 == 0) & (Z%2 != 0)):
                a5 = -12
            
            B = (a1*A[j]) - (a2*(A[j]**(2/3))) - ((a3*(Z**2))/(A[j]**(1/3))) - ((a4*((A[j]-2*Z)**2))/A[j]) + (a5/(A[j]**(1/2)))
    
            EUN.append(B/A[j])
    
        for k in range(0, len(EUN)):
            if EUN[k] == np.max(EUN):
                A_max = A[k]
            else:
                continue
    
        print("Número Atómico:", Z)
        print("El número de masa con el cual el átomo tiene mayor energía de enlace por nucleón es de:", A_max)
        print("La mayor energía de enlace por nucleón de éste átomo es de:", np.max(EUN), "MeV \n")

        EUN_max.append(np.max(EUN))

    print("La mayor energía de enlace por nucleón es de:", np.max(EUN_max), "MeV.")
```

```python
FSM_B_d()
```
